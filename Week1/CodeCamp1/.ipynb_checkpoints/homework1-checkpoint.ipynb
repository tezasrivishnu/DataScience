```python
import requests 
from bs4 import BeautifulSoup

request = requests.get('http://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_1970')
pageBillboard = request.text
soup = BeautifulSoup(pageBillboard, 'html.parser')
pageTables = soup.find_all('table', 'wikitable')
tableRows = [row for row in pageTables[0].find_all('tr')]
print(tableRows)
# tableRows

#we have to get td values of each row
def tdValues(inp):
    return [each for each in inp.find_all('td')]

#iterating from 1st index because in zeroth index only yhe column headings will be present
tdVal = [tdValues(each) for each in tableRows[1:]]
# print(tdVal)
#converting into given dictionary format in homework description
topBillboard = [{'band_singer':each[2].a.string, 'ranking':each[0].string, 'title':each[1].a.string,
                'url':each[2].a['href']} for each in tdVal]
topBillboard
```

```python
#getting the same data for the years 1992 to 2015
urls = ['http://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_'+str(i) for i in range(1992, 2015)]
def getData(url):
    yearsData = {}
    for url in urls:
        request = requests.get(url)
        yearsData[url.split('_')[-1]] = request.text
    return yearsData

yearsData = getData(urls)

yearsData
```

```python
# #parse and clean data
# def parse_year(the_year, yeartext_dict):
#     soup = BeautifulSoup(yeartext_dict, 'html.parser')
#     pageTables = soup.find_all('table', 'wikitable')
    
#     tableRows = [row for row in pageTables[0].find_all('tr')]
#     # print(len(tableRows))
#     # tableRows

#     #we have to get td values of each row
#     def tdValues(inp):
#         return [each for each in inp.find_all('td')]

#     #iterating from 1st index because in zeroth index only yhe column headings will be present
#     tdVal = [tdValues(each) for each in tableRows[1:]]
# #     print(len(tdVal))
#     print(tdVal)
#     #converting into given dictionary format in homework description
#     topBillboard = [{'band_singer':each[1].a.string, 'title':each[0].string, 'url':each[1].a['href']} for each in tdVal]
#     return topBillboard

# #lets get the data as above for year 1999
# print(parse_year(1998,yearsData['1998']))

def parse_year(the_year, yeartext_dict):

    soup = BeautifulSoup(yearsData[str(the_year)], 'html.parser')
    tableData = soup.find_all('table', 'wikitable')
    rows = [row for row in tableData[0].find_all('tr')]
#     print(rows)
    yearinfo = [singleYearData(row) for row in rows[1:]]
    return yearinfo



def singleYearData(row):
    rowEle = [each for each in row.children]
    rowEle = list(filter(lambda x: x != '\n', rowEle))
    ranking = rowEle[0].string
    ranking = ranking.replace("\n","")
#     print(ranking)

    rowSingers = rowEle[2].find_all('a')
#     print(rowSingers)
    rowSinger = [each.string for each in rowSingers]
#     print(rowSingers)
    singersUrl = [each['href'] for each in rowSingers]
#     print(singersUrl)
    rowSongs = rowEle[1].find_all('a')
    songsUrl = [each['href'] for each in rowSongs]
    if songsUrl == []:
        songsUrl = [None]
    songs = [each.string for each in rowSongs]
    title = '/'.join(str(each) for each in songs)

    songInfo = {'band_singer': rowSinger, 'ranking': ranking, 'song': songs,
                   'songurl':songsUrl, 'titletext':title, 'url': singersUrl}
    return songInfo
# print(parse_year(2014,yearsData['2014']))
# print(yearinfo)
```
