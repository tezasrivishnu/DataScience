```python
#linear regression with boston housing data that is loaded from sklearn
import statsmodels.api as sm
from statsmodels.formula.api import ols
#ols means ordinary least squares
```

```python
#fitting linearregression using ols in statmodel
m = ols('Price ~ RM',bostonDT).fit()
m.summary()
```

```python
plt.scatter(bostonDT['Price'], m.fittedvalues)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted Prices")
```

```python
#fitting linearregression using sklearn
from sklearn.linear_model import LinearRegression
#we are dropping the price from bostonDT
X = bostonDT.drop('Price', axis = 1)
```

```python
lm = LinearRegression()
lm
```

```python
#by using linear regression object we can use
#1. fit()----->will fit a linear model
#2. predict---->preict Y with linear coefficients
#3.score--->wil retur r squared value
```

```python
#fitting a linear model
lm.fit(X, bostonDT.Price)
```

```python
#estimated intercept coefficients
lm.intercept_
```

```python
#estimated number of coefficients
lm.coef_
```

```python
#data frame for each column and the estimated coefficients
pd.DataFrame(zip(X.columns, lm.coef_), columns = ['features', 'estimatedCoefficients'])
```

```python
#predicted values
lm.predict(X)
```

```python
#histogram for the predicted values
plt.hist(lm.predict(X))
plt.title('Predicted Housing Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
plt.scatter(bostonDT.Price, lm.predict(X))
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted Prices")
```

```python
#residual sum of squares
np.sum((bostonDT.Price - lm.predict(X)) ** 2)
```
